```python
import numpy as np
import matplotlib.pyplot as plt

# LÃ¶sung fÃ¼r Teil (a)
def u_a(x, y):
    return np.sin(y - 2*x) * np.exp(-x)

# Create grid
x = np.linspace(-2, 4, 30)
y = np.linspace(-4, 4, 30)
X, Y = np.meshgrid(x, y)

# Calculate vector field U = (1, 2) for the PDE
U_a = np.ones_like(X)  # âx u component is 1
V_a = 2 * np.ones_like(Y)  # ây u component is 2

# Calculate solution values
Z_a = u_a(X, Y)

# Plot solution
plt.figure(figsize=(15, 5))

# 3D surface plot
from mpl_toolkits.mplot3d import Axes3D
ax1 = plt.subplot(131, projection='3d')
ax1.plot_surface(X, Y, Z_a, cmap='viridis', alpha=0.8)
ax1.set_title('(a) LÃ¶sung $u(x,y) = \sin(y-2x)e^{-x}$')
ax1.set_xlabel('x')
ax1.set_ylabel('y')
ax1.set_zlabel('u(x,y)')

# Contour plot
plt.subplot(132)
contour = plt.contourf(X, Y, Z_a, levels=20, cmap='viridis')
plt.colorbar(contour)
plt.title('(a) HÃ¶henlinien der LÃ¶sung')
plt.xlabel('x')
plt.ylabel('y')

# Vector field
plt.subplot(133)
plt.quiver(X, Y, U_a, V_a, scale=20, color='red')
plt.streamplot(X, Y, U_a, V_a, color='blue', linewidth=1, density=1.5)
plt.title('(a) Vektorfeld (1,2) mit Stromlinien')
plt.xlabel('x')
plt.ylabel('y')
plt.axis('equal')

plt.tight_layout()
plt.show()
```

```python
# LÃ¶sung fÃ¼r Teil (b)
def u_b(t, x):
    return np.cos(x * np.exp(-t)) + (x**2)/2 * (1 - np.exp(-2*t))

# Create grid
t = np.linspace(0, 2, 20)
x = np.linspace(-3, 3, 30)
T, X_mesh = np.meshgrid(t, x)

# Calculate vector field U = (1, x) for the PDE
U_b = np.ones_like(T)  # ât u component is 1
V_b = X_mesh  # âx u component is x

# Calculate solution values
Z_b = u_b(T, X_mesh)

# Plot solution
plt.figure(figsize=(15, 5))

# 3D surface plot
ax1 = plt.subplot(131, projection='3d')
ax1.plot_surface(T, X_mesh, Z_b, cmap='plasma', alpha=0.8)
ax1.set_title('(b) LÃ¶sung $u(t,x)$')
ax1.set_xlabel('t')
ax1.set_ylabel('x')
ax1.set_zlabel('u(t,x)')

# Contour plot
plt.subplot(132)
contour = plt.contourf(T, X_mesh, Z_b, levels=20, cmap='plasma')
plt.colorbar(contour)
plt.title('(b) HÃ¶henlinien der LÃ¶sung')
plt.xlabel('t')
plt.ylabel('x')

# Vector field
plt.subplot(133)
plt.quiver(T, X_mesh, U_b, V_b, scale=20, color='red')
plt.streamplot(T, X_mesh, U_b, V_b, color='blue', linewidth=1, density=1.5)
plt.title('(b) Vektorfeld (1,x) mit Stromlinien')
plt.xlabel('t')
plt.ylabel('x')

plt.tight_layout()
plt.show()
```

```python
import numpy as np
import matplotlib.pyplot as plt
from mpl_toolkits.mplot3d import Axes3D

x_vals = np.linspace(-2, 2, 50)
y_vals = np.linspace(0.1, 3, 50)
X, Y = np.meshgrid(x_vals, y_vals)
U = X - Y + 2 / Y

fig = plt.figure()
ax = fig.add_subplot(111, projection='3d')
ax.plot_surface(X, Y, U, cmap='viridis', alpha=0.9)
ax.set_xlabel('x')
ax.set_ylabel('y')
ax.set_zlabel('u(x,y)')
ax.set_title(r'$u(x,y) = x - y + \frac{2}{y}$ fÃ¼r $y>0$')
plt.show()
```

```python
import numpy as np
import matplotlib.pyplot as plt
from matplotlib import rcParams

# Setze schÃ¶ne Schriftart
rcParams['font.family'] = 'serif'
rcParams['font.size'] = 12

# Definiere die Anfangsfunktion g(r)
def g(r):
    conditions = [
        r <= -1,
        (r > -1) & (r < 0),
        (r >= 0) & (r < 1),
        r >= 1
    ]
    values = [
        0,
        r + 1,
        -r + 1,
        0
    ]
    return np.piecewise(r, conditions, values)

# Erzeuge r-Werte fÃ¼r die Charakteristiken
r_values = np.concatenate([
    np.linspace(-2, -1, 5),      # Bereich 1
    np.linspace(-0.99, -0.01, 15), # Bereich 2
    np.linspace(0, 0.99, 15),    # Bereich 3
    np.linspace(1, 2, 5)         # Bereich 4
])

# Zeitbereich
t_max = 2
t = np.linspace(0, t_max, 100)

# Erstelle den Plot
plt.figure(figsize=(12, 8))

# Plotte Charakteristiken fÃ¼r jeden r-Wert
for r in r_values:
    x = g(r) * t + r
    # Unterschiedliche Farben fÃ¼r verschiedene Bereiche
    if r <= -1:
        color = 'blue'
        label = 'Bereich 1 (r â¤ -1)' if r == -2 else ""
    elif r < 0:
        color = 'red' 
        label = 'Bereich 2 (-1 < r < 0)' if r == -0.8 else ""
    elif r < 1:
        color = 'green'
        label = 'Bereich 3 (0 â¤ r < 1)' if r == 0.2 else ""
    else:
        color = 'purple'
        label = 'Bereich 4 (r â¥ 1)' if r == 1.2 else ""
    
    plt.plot(t, x, color=color, alpha=0.7, linewidth=1.5, label=label)

# Markiere wichtige Punkte
plt.axvline(x=1, color='black', linestyle='--', alpha=0.5, label='Schockzeit t=1')
plt.axhline(y=-1, color='gray', linestyle=':', alpha=0.5)
plt.axhline(y=0, color='gray', linestyle=':', alpha=0.5) 
plt.axhline(y=1, color='gray', linestyle=':', alpha=0.5)

# Beschriftung
plt.xlabel('Zeit t', fontsize=14)
plt.ylabel('Ort x', fontsize=14)
plt.title('Charakteristiken der Burgers-Gleichung', fontsize=16)
plt.grid(True, alpha=0.3)
plt.legend()

# Zeige den Bereich, wo sich Charakteristiken schneiden
plt.xlim(0, t_max)
plt.ylim(-2, 3)

plt.tight_layout()
plt.show()

# ZusÃ¤tzlich: Plot der Anfangsfunktion
plt.figure(figsize=(10, 6))
r_plot = np.linspace(-2, 2, 1000)
plt.plot(r_plot, g(r_plot), 'b-', linewidth=2, label='Anfangsbedingung g(r)')
plt.axvline(x=-1, color='red', linestyle='--', alpha=0.7, label='Grenzen')
plt.axvline(x=0, color='red', linestyle='--', alpha=0.7)
plt.axvline(x=1, color='red', linestyle='--', alpha=0.7)
plt.xlabel('r', fontsize=14)
plt.ylabel('g(r)', fontsize=14)
plt.title('Anfangsbedingung', fontsize=16)
plt.grid(True, alpha=0.3)
plt.legend()
plt.tight_layout()
plt.show()
```

```python
import numpy as np
import matplotlib.pyplot as plt

# Definiere die Anfangsfunktion g(r)
def g(r):
    conditions = [
        r <= -1,
        (r > -1) & (r < 0),
        (r >= 0) & (r < 1),
        r >= 1
    ]
    values = [
        0,
        r + 1,
        -r + 1,
        0
    ]
    return np.piecewise(r, conditions, values)

# Zeitbereich
t = np.linspace(0, 2, 200)

# WICHTIGE CHARAKTERISTIKEN:
plt.figure(figsize=(12, 8))

# 1. Linke Grenze: r = -1
r = -1
x_left = g(r) * t + r  # = 0*t -1 = -1
plt.plot(t, x_left, 'k-', linewidth=3, label='r = -1 (linke Grenze)')

# 2. Mitte links: r = -0.5 (Beispiel aus Bereich 2)
r = -0.5  
x_mid_left = g(r) * t + r  # = 0.5*t - 0.5
plt.plot(t, x_mid_left, 'r-', linewidth=2, label='r = -0.5 (Bereich 2)')

# 3. SPRUNGE: r = 0 von links und rechts
# r = 0 von links (gehÃ¶rt eigentlich zu Bereich 2)
r_left = 0
x_center_left = g(r_left) * t + r_left  # = 1*t + 0 = t
plt.plot(t, x_center_left, 'b-', linewidth=3, label='r = 0 (Mitte)')

# 4. Mitte rechts: r = 0.5 (Beispiel aus Bereich 3)  
r = 0.5
x_mid_right = g(r) * t + r  # = 0.5*t + 0.5
plt.plot(t, x_mid_right, 'g-', linewidth=2, label='r = 0.5 (Bereich 3)')

# 5. Rechte Grenze: r = 1
r = 1
x_right = g(r) * t + r  # = 0*t + 1 = 1
plt.plot(t, x_right, 'k-', linewidth=3, label='r = 1 (rechte Grenze)')

# 6. SCHOCK-LINIE (ab t=1) - wo sich Charakteristiken treffen
# FÃ¼r t >= 1: Schock bei x = 0.5*(1 + t) (vereinfacht)
t_shock = np.linspace(1, 2, 100)
x_shock = 0.5 * (1 + t_shock)  # Vereinfachte Schock-Position
plt.plot(t_shock, x_shock, 'm--', linewidth=4, label='Schock (ab t=1)')

# Markierungen
plt.axvline(x=1, color='red', linestyle=':', alpha=0.7, label='Schockzeit t=1')
plt.axhline(y=-1, color='gray', linestyle=':', alpha=0.5)
plt.axhline(y=0, color='gray', linestyle=':', alpha=0.5)
plt.axhline(y=1, color='gray', linestyle=':', alpha=0.5)

# Beschriftung
plt.xlabel('Zeit t', fontsize=14)
plt.ylabel('Ort x', fontsize=14)
plt.title('WICHTIGSTE CHARAKTERISTIKEN der Burgers-Gleichung', fontsize=16)
plt.grid(True, alpha=0.3)
plt.legend()
plt.xlim(0, 2)
plt.ylim(-1.5, 2.5)
plt.tight_layout()
plt.show()
```

```python
import numpy as np
import matplotlib.pyplot as plt
from mpl_toolkits.mplot3d import Axes3D

# Parametrisierung F1: u in [-2,2], v in (0, 2Ï)
u = np.linspace(-2, 2, 50)
v = np.linspace(0.001, 2*np.pi - 0.001, 50)  # Vermeide v=0 und v=2Ï
U, V = np.meshgrid(u, v)

# Katenoid-Parametrisierung
X = np.cosh(U) * np.cos(V)
Y = np.cosh(U) * np.cos(V)
Z = U

# Korrektur: Y sollte mit sin(v) berechnet werden
Y = np.cosh(U) * np.sin(V)

# 3D-Plot
fig = plt.figure(figsize=(10, 8))
ax = fig.add_subplot(111, projection='3d')

# OberflÃ¤che plotten
surf = ax.plot_surface(X, Y, Z, cmap='viridis', alpha=0.8, linewidth=0)

# Einige Parameterkurven hervorheben
# u = konstant (parallele Kreise)
for u_val in [-1.5, 0, 1.5]:
    v_line = np.linspace(0, 2*np.pi, 100)
    x_line = np.cosh(u_val) * np.cos(v_line)
    y_line = np.cosh(u_val) * np.sin(v_line)
    z_line = np.full_like(v_line, u_val)
    ax.plot(x_line, y_line, z_line, 'r-', linewidth=2)

# v = konstant (Meridiane)
for v_val in [0, np.pi/2, np.pi, 3*np.pi/2]:
    u_line = np.linspace(-2, 2, 100)
    x_line = np.cosh(u_line) * np.cos(v_val)
    y_line = np.cosh(u_line) * np.sin(v_val)
    z_line = u_line
    ax.plot(x_line, y_line, z_line, 'b-', linewidth=2)

ax.set_xlabel('X')
ax.set_ylabel('Y')
ax.set_zlabel('Z')
ax.set_title('Katenoid - Parametrisierung F1: v â (0, 2Ï)')
ax.set_box_aspect([1,1,1])

plt.show()

# Zeige die Definitionsbereiche
print("Parametrisierung F1: U = â Ã (0, 2Ï)")
print("Parametrisierung F2: U = â Ã (-Ï, Ï)")
print("Beide zusammen Ã¼berdecken den gesamten Katenoid.")
```
